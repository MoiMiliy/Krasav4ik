## 1 Від'ємний біноміальний розподіл

Більш детальна інформація в [документації](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.nbinom.html).

```python
from scipy.stats import nbinom
```

```python
n = 5
p = 0.4
```

### 1.1 Щільність

```python
x = np.arange(-0.5, n*5 + 1, 0.5)
```

Виведіть значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
nbinom_pmf = nbinom.pmf(x, n, p)
nbinom_pmf
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.stem(nbinom_pmf)
plt.show()
```

Зобразіть графічно значення функції щільності від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x1$.

```python
x1 = np.arange(0, n*5+1)
```

```python
plt.stem(nbinom.pmf(x1, n, p))
plt.show()

```

### 1.2 Функція розподілу

Виведіть значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
nbinom_cdf = nbinom.cdf(x, n, p)
nbinom_cdf
```

Зобразіть графічно значення функції розподілу від'ємного біноміального розподілу з параметрами $n, p$ в точках вектора $x$.

```python
plt.plot(nbinom_cdf)
plt.show()

```

### 1.3 Випадкові величини

Згенеруйте $100$ випадкових величин з від'ємного біноміального розподілу з параметрами $n, p$.

```python
nbinom_rvs = nbinom.rvs(n, p, size = 100)
nbinom_rvs
```

Виведіть кількість кожного елемента в отриманому випадковому векторі.

```python
for i in set(nbinom_rvs):
    print('Элемент {} встретился {} раз'.format(i, list(nbinom_rvs).count(i)))

```

Зобразіть графічно кількість кожного елемента в отриманому випадковому векторі.

```python
plt.hist(nbinom_rvs)
plt.show()
```

## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom

for n, p in [(15, 0.8), (15, 0.2), (20, 0.4), (20, 0.6), (7, 0.5), (20, 0.5), (100, 0.5)]:
    x = np.arange(0, n + 1)
    fig = plt.figure(figsize = (25, 7))
    plt.subplot(1, 3, 1, xlim = (0, n + 1))
    plt.stem(x, binom.pmf(x, n, p))
    plt.subplot(1, 3, 2, xlim = (0, n + 1))
    plt.plot(x, binom.cdf(x, n, p))
    plt.title('Для n = {}, p = {}'.format(n, p)) 
    plt.subplot(1, 3, 3, xlim = (0, n + 1))
    plt.hist(binom.rvs(n, p, size = 10000), normed = True, bins = 30)
    plt.show()

```

### 3.2 Геометричний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин геометричного розподілу з параметрами:
1. $p = 0.1$
2. $p = 0.2$
3. $p = 0.3$
4. $p = 0.4$
5. $p = 0.5$
6. $p = 0.6$
7. $p = 0.7$
8. $p = 0.8$
9. $p = 0.9$

```python
from scipy.stats import geom

A = np.arange(0.1, 1, 0.1)
for p in A:
    x = np.arange(0, 11)
    fig = plt.figure(figsize = (25, 7))
    plt.subplot(1, 3, 1, xlim = (0, 11))
    plt.stem(x, geom.pmf(x, p))
    plt.subplot(1, 3, 2, xlim = (0, 11))
    plt.plot(x, geom.cdf(x, p))
    plt.title('Для p = {}'.format(p))
    plt.subplot(1, 3, 3, xlim = (0, 11))
    plt.hist(geom.rvs(p, size = 10000), normed = True, bins = 30)
    plt.show()
```

### 3.3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
from scipy.stats import poisson

A = np.arange(0.1, 1, 0.1)
for u in A:
    x = np.arange(0, 11)
    fig = plt.figure(figsize = (25, 7))
    plt.subplot(1, 3, 1, xlim = (-0.1, 11))
    plt.stem(x, poisson.pmf(x, u))
    plt.subplot(1, 3, 2, xlim = (-0.1, 11)) 
    plt.plot(x, poisson.cdf(x, u))
    plt.title('Для u = {}'.format(u))
    plt.subplot(1, 3, 3, xlim = (-0.1, 11))
    plt.hist(poisson.rvs(u, size = 10000), normed = True)
    plt.show()

```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
n = 100

for p in [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]:
    x = np.arange(0, n + 1)
    fig = plt.figure(figsize = (10, 7))
    plt.stem(x, binom.pmf(x, n, p), 'r', markerfmt = 'ro', label = 'Биномиальное') 
    plt.stem(x, poisson.pmf(x, p*n), 'b', markerfmt = 'bo', label = 'Пуассона')
    plt.legend()
    plt.show()
```
